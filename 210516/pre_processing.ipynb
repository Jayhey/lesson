```python
# \n -> space로 변경 
# \x0 -> space로 변경 

test = df.abstract[0]
test = test.replace('\n', ' ')
test = test.replace('\xa0', ' ')
```

```python

# 함수로 만들어서 적용
def del_enter(x):
    x = x.replace('\n', ' ')
    x = x.replace('\xa0', ' ')
    return x

df.abstract = df.abstract.map(del_enter)
df.abstract[0]
```

```python
# PURPOSE, OBJECTIVE 등 제거
check_abstract = df[df.abstract.str.contains('PURPOSE')]
check_abstract.abstract[1]
```
