```python
# \n -> space로 변경 
# \x0 -> space로 변경 

test = df.abstract[0]
test = test.replace('\n', ' ')
test = test.replace('\xa0', ' ')
```

```python

# 함수로 만들어서 적용
def del_enter(x):
    x = x.replace('\n', ' ')
    x = x.replace('\xa0', ' ')
    return x

df.abstract = df.abstract.map(del_enter)
df.abstract[0]
```

```python
# 제거해야할 단어 목록
# PURPOSE:, METHODS:, RESULTS:, CONCLUSION: 
# PURPOSE OF REVIEW:, RECENT FINDINGS:
# PURPOSE/OBJECTIVES: , CONCLUSIONS:
# BACKGROUND/PURPOSE: 
# BACKGROUND:, DESIGN/METHODS:, 
# BACKGROUND AND PURPOSE:, EXPERIMENTAL APPROACH:, KEY RESULTS:, CONCLUSIONS AND IMPLICATIONS:
# STATEMENT OF PROBLEM:, MATERIAL AND METHODS:, 

del_words = [
'PURPOSE OF REVIEW: ',
'RECENT FINDINGS: ',
'PURPOSE/OBJECTIVES: ',
'BACKGROUND/PURPOSE: ',
'DESIGN/METHODS: ',
'BACKGROUND AND PURPOSE: ',
'EXPERIMENTAL APPROACH: ',
'KEY RESULTS: ',
'CONCLUSIONS AND IMPLICATIONS: ',
'STATEMENT OF PROBLEM: ',
'MATERIAL AND METHODS: ',
'Materials and methods:',
'LINKED ARTICLES: ',
'PURPOSE: ',
'METHODS: ',
'RESULTS: ',
'CONCLUSION: ',
'CONCLUSIONS: ',
'BACKGROUND: ']

def replace_words(x):
    for word in del_words:
        x = x.replace(word, '')
    return x


df.abstract = df.abstract.map(replace_words)
```
