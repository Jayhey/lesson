```python
import nltk, re, pprint
from nltk import word_tokenize
```

```python
test = df.abstract[0]
test
```

```python
# 일단 tokenize. 
text = word_tokenize(test)
text
```

```python
# stopwords 제거하기
from nltk.corpus import stopwords  
stop_words = set(stopwords.words('english')) 

del_stopwords = []
for w in text: 
    if w not in stop_words: 
        del_stopwords.append(w) 
```

```python
pos_tag_text = nltk.pos_tag(del_stopwords)
pos_tag_text
```

```python
want_words = []
for token in pos_tag_text:
    if token[1].startswith("V") or token[1].startswith("N") or token[1].startswith("JJ"):
        want_words.append(token[0])
```

```python
want_words
```

```python
def pos_tag_text(x):
    text = word_tokenize(x)
    
    del_stopwords = []
    for w in text: 
        if w not in stop_words: 
            del_stopwords.append(w) 
    
    pos_tag_text = nltk.pos_tag(del_stopwords)
    result = []
    for token in pos_tag_text:
        if token[1].startswith("V") or token[1].startswith("N") or token[1].startswith("JJ"):
            result.append(token[0])
    
    return result
```

```python
df.pos_tag = df.abstract.map(pos_tag_text)
```

```python
print(df.pos_tag[1])
```

```python
from gensim import corpora
```

```python
dictionary = corpora.Dictionary(df.pos_tag)
```

```python
corpus = [dictionary.doc2bow(text) for text in df.pos_tag]
```

```python
import gensim
NUM_TOPICS = 10
model = gensim.models.ldamodel.LdaModel(corpus, 
                                        num_topics=NUM_TOPICS, 
                                        id2word=dictionary, 
                                        passes=15)
```
