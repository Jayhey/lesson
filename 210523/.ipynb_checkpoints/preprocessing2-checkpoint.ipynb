```python
# 추가 제거 단어 목록 추가

del_words = [
    'PURPOSE OF REVIEW: ',
    'RECENT FINDINGS: ',
    'PURPOSE/OBJECTIVES: ',
    'BACKGROUND/PURPOSE: ',
    'DESIGN/METHODS: ',
    'BACKGROUND AND PURPOSE: ',
    'EXPERIMENTAL APPROACH: ',
    'KEY RESULTS: ',
    'CONCLUSIONS AND IMPLICATIONS: ',
    'STATEMENT OF PROBLEM: ',
    'MATERIAL AND METHODS: ',
    'MATERIALS AND METHODS:',
    'LINKED ARTICLES: ',
    'PURPOSE: ',
    'METHODS: ',
    'RESULTS: ',
    'CONCLUSION: ',
    'Conclusion: ',
    'CONCLUSIONS: ',
    'Conclusions: ',
    'BACKGROUND: ', 
    'Background: ',
    'AIMS: ',
    'KEY FINDINGS: ',
    'SIGNIFICANCE: ',
    'OBJECTIVES: ']

def replace_words(x):
    for word in del_words:
        x = x.replace(word, '')
    return x


df.abstract = df.abstract.map(replace_words)
```

```python
# 이상한 year 값 들어 있으면 제외하기 
del_year = ['Summ', 'None', 'F100']

df = df[~df.year.str.contains('|'.join(del_year))]
```

```python
import nltk, re, pprint
from nltk import word_tokenize
```

```python
test = df.abstract[0]
test
```

```python
# 일단 tokenize. 
text = word_tokenize(test)
text
```

```python
# stopwords 제거하기
from nltk.corpus import stopwords  
stop_words = set(stopwords.words('english')) 

del_stopwords = []
for w in text: 
    if w not in stop_words: 
        del_stopwords.append(w) 
```

```python
pos_tag_text = nltk.pos_tag(del_stopwords)
pos_tag_text
```

```python
want_words = []
for token in pos_tag_text:
    if token[1].startswith("V") or token[1].startswith("N") or token[1].startswith("JJ"):
        want_words.append(token[0])
```

```python
want_words
```

```python
def pos_tag_text(x):
    text = word_tokenize(x)
    
#     del_stopwords = []
#     for w in text: 
#         if w not in stop_words: 
#             del_stopwords.append(w) 
    
    pos_tag_text = nltk.pos_tag(del_stopwords)
    result = []
    for token in pos_tag_text:
        if token[1].startswith("V") or token[1].startswith("N") or token[1].startswith("JJ"):
            result.append(token[0])
    
    return result
```
