```python
import pandas as pd
from gensim import corpora
from nltk import word_tokenize, pos_tag
```

```python
df = pd.read_csv("/Users/geoffrey/jupyter/lesson/210124/to_210128.csv")
```

```python
df
```

```python
# 동일한 표현 : df[df['abstracts'] != 'Error']
docs = df[df.abstracts != 'Error'].abstracts
```

```python
# 형용사, 명사, 고유명사를 뽑는 함수 만들어주기
def extract_pos(x):
    tokenized = word_tokenize(x)
    tmp = []
    pos_taged = pos_tag(tokenized)
    for (word, pos) in pos_taged:
        if pos in ['NN', 'JJ', 'NNP']:
            tmp.append(word)
    
    return tmp
```

```python
docs = docs.map(extract_pos)
```

```python
dictionary = corpora.Dictionary(docs)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
print(corpus[1])
```

```python
# lda_model, LdaModel : snake_case, CamelCase
from gensim.models.ldamodel import LdaModel

NUM_TOPICS = 5
ldamodel = LdaModel(corpus, num_topics=NUM_TOPICS, id2word=dictionary, passes=15)
```

```python
# A: 'apple', 'banana', 'dog'
# B: 'apple', 'pinepple', 'lemon', '...'
# C: 'apple', 'cat', 'house', "...."

# 'apple', 'banana', 'cat', 'dog', 'house', 'lemon', 'pineapple', '...'


# TOPIC : T1, T2, T3 ... T5
# Word, Topic
# Document, Topic


# 알 수 있는 것 1
# 문서가 어떤 토픽에 얼만큼 분포되어 있는가 (토픽의 분포)
# Doc A : T1 0.05, T2 0.20, T3 0.00, ... T5 0.70    Sum = 1
# Doc B : T1 0.30, T2 0.20, T3 0.10, ... T5 0.20

# 알 수 있는 것 2
# 토픽에서 중요한 단어들
# T1 : Appple 0.05, banana 0.70, ... 
# T2 : Cat 0.80, Dog 0.05, ... (동물?)
# T5 : Apple 0.20, banana 0.40, dog 0.30, ... (동물과 과일인가?)
```

```python
# 알 수 있는 것 1
for i, topic_list in enumerate(ldamodel[corpus]):
    print(f"{i}번째 문서의 토픽 분포는 {topic_list}")
```

```python
# 알 수 있는 것 2
topics = ldamodel.print_topics(num_words=20)

for topic in topics:
    print(topic)
```
