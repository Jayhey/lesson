```python
# 우리가 계산기를 만들고 싶다면?

result = 0

def add(num):
    global result
    result += num
    return result
```

```python
# 너무 많은 계산기를 만들기 어렵다면 Class 활용

class Cal:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result += num
        return self.result
    
    def multiply(self, num):
        self.result *= num
        return self.result
    
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(test_title)
```

```python

words = []
for abstract in list(df.abstracts):
    tmp = tokenizer.tokenize(abstract)
    words.append(tmp)
```

```python
words = [x for x in words if x != ['Error']]

# 아래는 같은 표현식 
# test = []
# for x in words:
#     if x != ['Error']:
#         test.append(x)
```

```python
words = list(map(lambda x: ' '.join(x), words))
```

```python
text = ' '.join(words)
text
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline

stopwords = set(STOPWORDS)

custom_stop_words = ['group', 'used', 'using', 'showed',
                     'objective', 'result', 'results', 'patient', 'patients', 'method', 'concolusion', 'Conclusion']

for stop_word in custom_stop_words:
    stopwords.add(stop_word)
```

```python
wordcloud = WordCloud(max_font_size=50, stopwords=stopwords).generate(text)
```

```python

fig = plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
