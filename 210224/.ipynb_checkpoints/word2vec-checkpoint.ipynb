## 실제 abstracts를 가지고 진행해보기

```python
# tokenize 복습!
# tokenize : 문장을 token (학습에 사용되는 최소 단위) 으로 쪼개는 작업
df.abstracts[0]
```

```python
test = df.abstracts[0]
print(word_tokenize(test))
```

```python
df.abstracts
```

```python
# series.map(function) -> 시리즈의 row에 함수를 전부 적용한다
# 이미 쪼개진 단어들에 대해서 pos_tag를 진행

def extract_pos(x):
    tokenized = word_tokenize(x)
    tmp = []
    pos_taged = pos_tag(tokenized)
    for (word, pos) in pos_taged:
        if pos in ['NN', 'JJ', 'NNP']:
            tmp.append(word)
    
    return tmp

df.abstracts = df.abstracts.map(extract_pos)
```

```python
# word2vec의 input : list of list  (ex :  [['a','b','c'], ['a','d','f'], ...]  )
# numpy 기준으로는 2d array 라고 함
# pandas series를 list로 바꾸는 방법 : series.tolist()
docs = df.abstracts.tolist()
print(docs[2:4])
```

```python
%%time
# %%time을 치면 얼마나 걸렸는지 표시해줌
# word2vec 학습
model = Word2Vec(sentences=docs,
                 window=5, # window size (양옆 어느정도 거리까지 학습할 것인지)
                 min_count=1, # 최소 이정도는 등장해야 학습에 포함시킴
                 workers=4, # 사용할 cpu threads
                 sg=1, # 1이면 skip-gram, 0이면 C-BOW
                 size=50, # 공식 홈페이지 parameter 기준 : vector_size 
                 iter=200, # 공식 홈페이지 parameter 기준 : epochs
                )


# vector_size, epochs 하이퍼파라미터 안먹는 이유 알아보기 
```

```python
# window size 3
model.wv.most_similar(positive=["orthodontic"])
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
# 시각화 툴  (더 쉬운 seaborn 이라는 패키지고 있음!)
from matplotlib import pyplot as plt
```

```python
# word2vec에서 학습된 모든 단어들
vocab = list(model.wv.vocab)
```

```python
# deprecate warning : 안 쓰는게 좋음
model[vocab]
```

```python
# deprecate warning에서 대체법을 알려준대로 사용
X = model.wv.__getitem__(vocab)
```

```python
size = 200
X_show = X[:size,:]
vocab_show = vocab[:size]
X_show.shape
```

```python
# 방법 1 : pca
pca = PCA(n_components=2)

pca.fit(X_show)

x_pca = pca.transform(X_show)
```

```python
# 400 * 2 짜리로 pca 적용
x_pca.shape
```

```python
# 그림의 사이즈
plt.figure(figsize=(30, 20))
# x축, y축 길이를 지정
plt.xlim(x_pca[:, 0].min(), x_pca[:, 0].max())
plt.ylim(x_pca[:, 1].min(), x_pca[:, 1].max())

for i in range(len(X_show)):
    plt.text(x_pca[i, 0], x_pca[i, 1], str(vocab_show[i]), fontdict={'weight': 'bold', 'size': 15})

plt.xlabel("pca1")
plt.ylabel("pca2")
plt.show()
```

```python
tsne = TSNE(n_components=2)
x_tsne = tsne.fit_transform(X_show)

```

```python

# 그림의 사이즈
plt.figure(figsize=(30, 20))
# x축, y축 길이를 지정
plt.xlim(x_tsne[:, 0].min(), x_tsne[:, 0].max())
plt.ylim(x_tsne[:, 1].min(), x_tsne[:, 1].max())

for i in range(len(X_show)):
    plt.text(x_tsne[i, 0], x_tsne[i, 1], str(vocab_show[i]), fontdict={'weight': 'bold', 'size': 15})

plt.xlabel("tsne1")
plt.ylabel("tsne2")
plt.show()
```
