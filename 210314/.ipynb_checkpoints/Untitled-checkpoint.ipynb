```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# toy dataset
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
# matplotlib: 조금 더 low-level plot tool
# seaborn: 한 줄로 만들 수 있는 high-level plot tool (matplotlib으로 만들어짐)
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.set_theme(style='darkgrid')
```

```python
# plot.show() 를 써주면 <seaborn.axisgrid.FacetGrid at 0x7fb09f65f110> 이런 메시지가 안나옴
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='time', style='smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', size='size', sizes=(15, 300), data=tips)
```
