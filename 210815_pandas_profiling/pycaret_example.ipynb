```python
 # 데이터 가져오기
 data = get_data('juice')
```

```python
# Data Setup

from pycaret.classification import *

clf = setup(data, # 대상 데이터. dataframe 형식
            target='Purchase', # 해당 dataframe의 target column
            # 만약 numeric, categorical을 수동으로 설정해줘야 한다면 아래처럼 지정
            # numeric_feature=['col1','col2',... ,'col'],
            # categorical_feature=['col5','col7',...,'col27'],
            log_experiment=True,
            experiment_name='juice_1')
```

```python
# 모델 비교하기 
best_model = compare_models()
```

```python
# 모델 생성하기
# https://pycaret.org/create-model/ 모델 이름은 이곳 참조
lr = create_model('lr')
```

```python
dt = create_model('dt')
```

```python
rf = create_model("rf", fold=5)
```

```python
models()
```

```python
# 앙상블 모델 목록을 보고싶을 때
models(type='ensemble')
```

```python
# 모델 튜닝
tuned_rf = tune_model(rf)
```

```python
# bagging 
bagged_dt = ensemble_model(dt, method='Bagging')
```

```python
boosted_dt = ensemble_model(dt, method='Boosting')
```

```python
# 모델 여러개를 섞고 싶을 때
blender = blend_models(estimator_list=[boosted_dt, bagged_dt, tuned_rf])
```
