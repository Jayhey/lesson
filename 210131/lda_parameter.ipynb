```python
# 'Error' 제외하기
df = df[df.abstracts != 'Error']
df = df[df.meta_keywords != 'Error']

# index reset 해주기
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# 필요 없는 문자 지워주기
def delete_colon_words(x):
    x = x.replace('Objective:', '')
    x = x.replace('Results:', '')
    x = x.replace('Methods:', '')
    x = x.replace('Conclusions:', '')
    return x

# 형용사, 명사, 고유명사를 뽑는 함수 만들어주기
def extract_pos(x):
    tokenized = word_tokenize(x)
    tmp = []
    pos_taged = pos_tag(tokenized)
    for (word, pos) in pos_taged:
        if pos in ['NN', 'JJ', 'NNP']:
            tmp.append(word)
    
    return tmp
```

```python
# pandas dataframe에는 map function을 사용할 수 없음! 
# df.map(extract_pos)  ->  (에러 발생)
# pandas series 형식에만 사용 가능함

df.abstracts = df.abstracts.map(delete_colon_words)
df.abstracts = df.abstracts.map(extract_pos)
```

```python
# 아래 두 인자를 받는 함수 선언
# abstract : abstracts 칼럼의 각각의 행
# meta_keyword : meta_keywords 칼럼의 각각의 행
def concat_abstracts_keywords(abstract, meta_keyword):

    meta_keyword = meta_keyword.replace('.', '')
    meta_keyword = meta_keyword.split(',')

    result = abstract + meta_keyword
    return result
```

### 정수 인코딩 및 lda 모델 돌려보기

```python
dictionary = corpora.Dictionary(docs)
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
%%time

# lda_model, LdaModel : snake_case, CamelCase
from gensim.models.ldamodel import LdaModel

NUM_TOPICS = 10
ldamodel = LdaModel(corpus=corpus,
                    id2word=dictionary,
                    num_topics=NUM_TOPICS,
                    
                    passes=50)
```

```python
# 알 수 있는 것 1
for i, topic_list in enumerate(ldamodel[corpus]):
    print(f"{i}번째 문서의 토픽 분포는 {topic_list}")
```

```python
# 알 수 있는 것 2
topics = ldamodel.print_topics(num_words=30)

for topic in topics:
    print(topic)
```

```python
# 필요 없는 단어를 제거하기 

def remove_words(doc):
    remove_list = ['group', 'significant', '<', 'p', 'study', '=', '%', 'mm']
    result = [x for x in doc if x not in remove_list]
    return result
```

```python
docs = docs.map(remove_words)
```

```python
# 단어사전만들기
dictionary = corpora.Dictionary(docs)

# A B C D E ... P


# 단어사전을 이용한 정수인코딩

#      A B C D E ... P
# doc1 1 0 1 1 0 ... 1
# doc2 ...
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
%%time

# 다시 한번 학습
from gensim.models.ldamodel import LdaModel

NUM_TOPICS = 5

# 
ldamodel = LdaModel(corpus=corpus,
                    id2word=dictionary,
                    num_topics=NUM_TOPICS, # 토픽의 수
                    chunksize=363, # 큰 변화 없음. 학습을 빠르게 하고 싶으면 높은 값을 주면 됨. (최대 문서의 수)
                    passes=50, # 많을 수록 깔끔한 결과가 나옴.
                    alpha='auto')
```

```python
# LDA를 평가하는 지표인 perplexity (사전적 의미 : 당혹감?)
# 작으면 작을 수록 좋음. 
ldamodel.log_perplexity(corpus)
```

```python
# 알 수 있는 것 1
for i, topic_list in enumerate(ldamodel[corpus]):
    print(f"{i}번째 문서의 토픽 분포는 {topic_list}")
```
