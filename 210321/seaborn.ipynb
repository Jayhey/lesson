```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins
```

```python
# NaN 데이터를 drop 해주는 명령어
# inplace=True : 바로 덮어쓰기를 해줌 (동일 : penguins = penguins.dropna())
penguins.dropna(inplace=True)
```

```python
# 현재 대상 데이터가 float 형태가 아니기 때문에 그래프를 그릴 수 없음
sns.distplot(penguins, x='flipper_length_mm')
```

```python
# check type for column
penguins.flipper_length_mm = penguins.flipper_length_mm.astype(float)
```

# Categorical data visualize

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

```python
# 해당 명령어에 대한 세부 내용을 볼 수 있음
??sns.catplot
```

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips)
# "strip", "swarm", "box", "violin",
#     "boxen", "point", "bar", or "count".
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

```python
sns.catplot(x='total_bill', y='day', hue='sex', kind='swarm', data=tips)
```

```python
# split parameter
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', split=True, data=tips)
```

```python
g = sns.catplot(x='day', y='total_bill', inner=None, kind='violin', data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.survived.value_counts()
```

```python
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=titanic)
```

```python
sns.catplot(x='class', y='survived', hue='sex',
           palette={"male": "g", "female": "m"},
           markers=['^', 'o'], linestyles=['-', '--'],
           kind='point', data=titanic)
```
